```python
## Returns a dictionary of match ids and the players from that match. Key: match_ids (String). Pair: (Player 1, Player 2) (Tuple)

def getPlayersFromIds(matches_fileloc):

    matches = pd.read_csv("../data/raw/match/2024-usopen-matches.csv")

    match_id_data = { }

    for _, row in matches.iterrows():
    
        p1 = row['player1']
        p2 = row['player2']
        id = row['match_id']

        match_id_data[id] = (p1, p2)

    return match_id_data
```
